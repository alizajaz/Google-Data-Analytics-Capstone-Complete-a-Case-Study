```python

# Concatenate the DataFrames along rows
data_frames = [jun04,jun05,july06,aug07]
 
merge_df = pd.concat(data_frames,ignore_index=True)
merge_df
```

```python
merge_df=merge_df.drop_duplicates()
merge_df
```

```python
# List of columns to be dropped
columns_to_drop=['start_lat','start_lng','end_lat','end_lng','start_station_name','start_station_id','end_station_name','end_station_id']
merge_df=merge_df.drop(columns=columns_to_drop)
merge_df
```

```python
#Check missing data in each column of the dataset
#this dataset doesn't have any missing values in the specified columns except start_station_name, start_station_id....these has been remove already

print(merge_df.isnull().sum())
```

```python
# Strip white spaces from all columns
merge_df = merge_df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
merge_df
```

```python
# Convert time columns to datetime objects
merge_df['started_at'] = pd.to_datetime(merge_df['started_at'])
merge_df['ended_at'] = pd.to_datetime(merge_df['ended_at'])

# Subtract 'c2' from 'd2'
merge_df['formatted_time'] = merge_df['ended_at'] - merge_df['started_at']
merge_df

```

```python
# # # Convert timedelta column back to 'hhmmss' format
# merge_df['ride_length'] = merge_df['ride_length'].astype(str)
# merge_df

 #Extract days and time into separate columns
# Calculate hours, minutes, and seconds
# Extract HH:MM:SS components

merge_df['hours'] = merge_df['formatted_time'].dt.components['hours']
merge_df['minutes'] = merge_df['formatted_time'].dt.components['minutes']
merge_df['seconds'] = merge_df['formatted_time'].dt.components['seconds']

# Create a new column with 'HH:MM:SS' format

merge_df['ride_length'] = merge_df['hours'].astype(str).str.zfill(2) + ':' +merge_df['minutes'].astype(str).str.zfill(2) + ':' + merge_df['seconds'].astype(str).str.zfill(2)

merge_df





```

```python
merge_df.dtypes
```

```python
# Create a new column 'weekday' and insert the weekday for 'time_started'
merge_df['weekday'] = merge_df['started_at'].dt.strftime('%A')
merge_df
```

```python
# Creating a new column 'total_seconds'
merge_df['total_seconds'] = merge_df['hours']*3600 + merge_df['minutes']*60 + merge_df['seconds']
merge_df
# mean_minute = merge_df['minutes'].mean()
# mean_minute
```

```python
# Calculate the mean for the 'total_seconds' column and assign it to a new column 'average_ride_length'
merge_df['average_ride_length'] = merge_df['total_seconds'].mean()

# Calculate the mean for each row and assign it to a new column 'average_ride_length'
merge_df['average_ride_length'] = merge_df.mean(axis=1)
merge_df
```
